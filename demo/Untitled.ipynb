```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
r_min = 30
r_max = 120
```

```python
fname = "/home/adam/Pictures/index2.png"
```

```python
def show(img):
    plt.figure(figsize = (10,10))
    plt.imshow(img[:,:,::-1])
    plt.show()
```

```python
img = cv2.imread(fname, cv2.IMREAD_COLOR)
show(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 2, r_min,
                           minRadius = r_min, maxRadius = r_max)
print(circles)
```

```python
circles.shape
```

```python
circles[0, :, :]
```

```python
scratch = img.copy()
```

```python
color = (0, 255, 0)
for (x, y, rad) in circles[0, :, :]:
    print(x,y)
    cv2.circle(scratch, (int(x), int(y)), int(rad), color, -1)
show(scratch)
```
